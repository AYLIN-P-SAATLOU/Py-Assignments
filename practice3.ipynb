```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Loop through primes <= 100
for p in range(2, 101):
    if is_prime(p):
        candidate = 2**p - 1
        if is_prime(candidate):
            print(f"2^{p} - 1 = {candidate} is prime")
        else:
            print(f"2^{p} - 1 = {candidate} is NOT prime")
```
